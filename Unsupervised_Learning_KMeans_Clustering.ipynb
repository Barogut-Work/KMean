# Unsupervised Learning: KMeans Clustering (Iris Dataset)


```python
import pandas as  pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
%config InlineBackend.figure_format='retina'
```

```python
print(f'pandas version = {pd.__version__}')
print(f'numpy version = {np.__version__}')
print(f'seaborn version = {sns.__version__}')
```

```python
df=pd.read_csv('https://github.com/prasertcbs/basic-dataset/raw/master/iris.csv')
df.head(10)
```

```python
df.species.value_counts()
```

```python
df.columns
```

```python
# สร้าง Pair Plot เพื่อแสดงความสัมพันธ์ระหว่างตัวแปรต่างๆ
# โดยแบ่งกลุ่มข้อมูลตาม species และใช้ markers ต่างกัน
sns.pairplot(df , vars=['sepal_length',  'sepal_width',  'petal_length',  'petal_width'],  # ฟังก์ชันสร้าง Pair Plot ใน Seaborn vars= ตัวแปรที่ต้องการพล็อต (ขนาดของกลีบเลี้ยงและกลีบดอก)
             hue='species',                                  # แบ่งกลุ่มข้อมูลตาม species และกำหนดสี
             markers=['o' , 'D' , 's' ] ,                  # ใช้ markers 'o', 'D', 's' ตามลำดับของ species
             plot_kws={'alpha':  .4});                # ตั้งค่าความโปร่งใสของ markers เป็น 0.4
```
